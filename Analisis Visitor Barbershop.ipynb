```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the dataset
goo_bisnis = pd.read_csv('./data_googlebusiness.csv')[1:]
goo_bisnis.head()
```

### Keterangan:
* Overall rating: Overall rating of this listing as of the date the report was generated
* Total searches: Number of times customers found this listing by searching on Google Search or Maps
* Direct searches: Number of times customers found this listing by searching for the business name or address on Google Search or Maps
* Discovery searches: Number of times customers found this listing by searching for the category, product or service on Google Search or Maps
* Total views: Number of times this listing has been viewed on Google Search or Maps,Number of times this listing has been viewed on Google Search
* Maps views: Number of times this listing has been viewed on Google Maps
* Total actions: Number of times customers have taken action on this listing on Google Search or Maps
* Website actions: Number of times customers have visited the website from this listing
* Directions actions: Number of times customers have requested directions from this listing
* Phone call actions: Number of times customers have called the business from this listing

```python
columns = goo_bisnis.columns[3:]

for column in columns:
    goo_bisnis[column] = pd.to_numeric(goo_bisnis[column])
```

```python
%matplotlib inline
```

```python
# How many categories we want to see their trend?
num_cat = len(goo_bisnis.columns[3:]) # without Start Date, End Date, Overall Rating

# Make subplots for each category
for i in range(num_cat):
    plt.figure(figsize=(16,10))
    plt.subplot(5,2,i+1)
    plt.plot(goo_bisnis['Start Date'], goo_bisnis[goo_bisnis.columns[3:][i]])
    plt.xticks(goo_bisnis['Start Date'], rotation=90)
    plt.xlabel('Start Date', fontsize=10)
    plt.ylabel(goo_bisnis.columns[3:][i].capitalize(), fontsize=10)
    plt.title(goo_bisnis.columns[3:][i].capitalize(), fontsize=12)
    plt.grid()
```

Kesimpulan: Customer relatif jarang melakukan phone call, tetapi lebih sering melalui pencarian online. Umumnya, melalui pencarian via Maps ataupun searching.

 * Alasan: Trendnya mulai meningkat pada minggu-minggu pertama. Meskipun terdapat titik puncak pada tiap trend, hal tersebut terjadi jarang sekali.
